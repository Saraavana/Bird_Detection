```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas_datareader
import datetime
import pandas_datareader.data as web
```

```python
start = datetime.datetime(2014,1,1)
end = datetime.datetime(2021,1,1)
```

```python
bmw = web.DataReader('BMWYY','yahoo',start,end)
bmw.head()
```

```python
tesla = web.DataReader('TSLA','yahoo',start,end)
tesla.head()
```

```python
# Data Visualization
# Linear plot of the open proce of BMW & Tesla stocks
bmw['Open'].plot(label='BMW',title='Open Prices',figsize=(12,8))
tesla['Open'].plot(label='Tesla')
plt.legend()
```

```python
# Let's plot the volume of stock that is traded each day
bmw['Volume'].plot(label='BMW',title='Volume Traded',figsize=(12,8))
tesla['Volume'].plot(label='Tesla')
plt.legend()
```

```python
# Calculating Total money traded = Volume traded * Open price
bmw['Total trade amount'] = bmw['Open']*bmw['Volume']
tesla['Total trade amount'] = tesla['Open']*tesla['Volume']
```

```python
bmw['Total trade amount'].plot(label='BMW', title="Total Amount Traded",figsize=(10,8))
tesla['Total trade amount'].plot(label='Tesla')
plt.legend()
```

```python
# Let's calculate when highest amount of trade happened with Tesla 
m1 = tesla['Total trade amount'].max()
tesla.loc[tesla['Total trade amount']==m1]
```

```python
# Let's calculate when highest amount of trade happened with BMW 
m1 = bmw['Total trade amount'].max()
bmw.loc[bmw['Total trade amount']==m1]
```

```python
# Financial analysis
# Daily percentage return: r(t) = (p(t)/p(t-1)) - 1;
# where r(t) is return at time t and p(t) is price at time t. 
# This helps in determining the percentage gain or loss and to analyze the volatility of the stock. 
# Volatility is the rate at which the price of a stock increases or decreases over a particular period.
# Hence, if the stock rises and falls more than 1% for a period of time then it is called as a volatile market. 
# When we plot a histogram and the distribution is wide then we can say its more volatile.
# For time (t-1), the 'shift' method is used to shift down/forward by period 1.
```

```python
bmw['returns'] = (bmw['Close']/bmw['Close'].shift(1))-1
tesla['returns'] = (tesla['Close']/tesla['Close'].shift(1))-1
```

```python
tesla['returns'].hist(bins=100, label='Tesla',figsize=(10,8))
bmw['returns'].hist(bins=100, label='BMW',figsize=(10,8))
plt.legend()
```

```python
# Looking at the histogram, we can say BMW has stable market but Tesla has high volatility because it has a wide distribution
```

```python
# Let's estimate kernel density estimation (KDE) plot for better insight
bmw['returns'].plot(kind='kde', label='BMW',figsize=(10,8))
tesla['returns'].plot(kind='kde', label='Tesla',figsize=(10,8))
plt.legend()
```

```python
# Let's plot a scatter plot matrix to compare the returns between BMW & Tesla stocks and analyze the correlation between them
```

```python
from pandas.plotting import scatter_matrix
```

```python
comp_df = pd.concat([bmw['returns'],tesla['returns']],axis=1)
comp_df.columns = ['BMW returns','Tesla returns']
scatter_matrix(comp_df,figsize=(8,8),hist_kwds={'bins':100})
```

```python
# Cumulative Return
# Cumulative means the aggregation of stock amount over a period of time. 
# While considering cumulative return, it is calculated according to the day the investment was made,
# unlike the daily return where it is calculated according to the previous day. 
# If the cumulative return is above 1, profit is generated otherwise it's a loss.
# Cumulative Return = i(i) = (1+r(t)).i(t-1) 
# where i(t) is investement at time t and r(t) is return at time t.
```

```python
bmw['cumulative return'] = (1 + bmw['returns']).cumprod()
tesla['cumulative return'] = (1 + tesla['returns']).cumprod()
```

```python
bmw['cumulative return'].plot(label='BMW')
tesla['cumulative return'].plot(label='Tesla')
plt.legend()
```

```python
# From the above plot we can assert that Tesla has largest cumulative return than BMW over time.
```
